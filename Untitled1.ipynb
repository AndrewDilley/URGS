```python
import pandas as pd

# Define the path to the CSV file. Using a raw string (r'...') avoids issues with backslashes.
file_path = r'C:\Users\andrew.dilley\development\URGS\DebtProfile.csv'

# Load the CSV data into a DataFrame
df = pd.read_csv(file_path)

# Inspect the first few rows of the DataFrame
print("First 5 rows of the dataset:")
print(df.head())

# Get a concise summary of the DataFrame, including the data types of each column
print("\nDataFrame information:")
df.info()

# Get summary statistics for numerical columns (if applicable)
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
# Assuming df is your DataFrame
print("Column Names:")
print(df.columns.tolist())
```

```python
# Find rows where pk_acount is duplicated
duplicate_rows = df[df['pk_acount'].duplicated(keep=False)]
print("Rows with duplicate pk_acount values:")
print(duplicate_rows.sort_values('pk_acount'))
```

```python
# Group by pk_account and aggregate:
# - Sum the amt_debit for each account and keep the column name as amt_debit
# - For the other columns, use 'first'
aggregated_df = df.groupby('pk_account', as_index=False).agg({
    'fk_financial_year': 'first',
    'month': 'first',
    'pk_porg': 'first',
    'pk_acount': 'first',
    'pension_type_desc': 'first',
    'amt_debit': 'sum',  # This column remains amt_debit after summing
    'town': 'first'
})

print("New DataFrame with one row per pk_account and aggregated amt_debit:")
print(aggregated_df)
```

```python
df = aggregated_df
```

```python
# Get all unique values in the pension_type_desc column
unique_pension_types = df['pension_type_desc'].unique()

# Print the unique values
print("Unique values in pension_type_desc:")
print(unique_pension_types)
```

```python
import matplotlib.pyplot as plt

# Create a histogram for the amt_debit column with 30 bins (adjust bins as needed)
plt.figure(figsize=(10, 6))
plt.hist(df['amt_debit'], bins=30, edgecolor='black')
plt.title('Distribution of amt_debit (Debt)')
plt.xlabel('amt_debit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate IQR and define bounds for outlier removal
Q1 = df['amt_debit'].quantile(0.25)
Q3 = df['amt_debit'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
filtered_data = df[(df['amt_debit'] >= lower_bound) & (df['amt_debit'] <= upper_bound)]

# Plot a histogram for the filtered amt_debit data with more bins for greater granularity
plt.figure(figsize=(10, 6))
plt.hist(filtered_data['amt_debit'], bins=50, edgecolor='black')  # Adjust bins as needed
plt.title('Distribution of amt_debit (Debt) without Outliers')
plt.xlabel('amt_debit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate IQR and define bounds for outlier removal
Q1 = df['amt_debit'].quantile(0.25)
Q3 = df['amt_debit'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers from amt_debit
filtered_data = df[(df['amt_debit'] >= lower_bound) & (df['amt_debit'] <= upper_bound)]

# Exclude rows where pension_type_desc is NaN
filtered_data = filtered_data.dropna(subset=['pension_type_desc'])

# Get unique pension types from the filtered data
pension_types = filtered_data['pension_type_desc'].unique()

# Create a list of arrays, each containing the amt_debit values for a specific pension type
data_by_pension = [
    filtered_data[filtered_data['pension_type_desc'] == pension]['amt_debit']
    for pension in pension_types
]

# Plot the stacked histogram with 50 bins for greater granularity
plt.figure(figsize=(10, 6))
plt.hist(data_by_pension, bins=50, stacked=True, edgecolor='black', label=pension_types)
plt.title('Stacked Distribution of Debt by Concession Type (No Outliers)')
plt.xlabel('Debt ($)')
plt.ylabel('Number of Customers')
plt.legend(title='Concession Type', loc='upper right')
plt.grid(True)
plt.show()
```
