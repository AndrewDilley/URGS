```python
import pandas as pd


# 2. Rename description columns
account_district.rename(columns={'district': 'district_description'}, inplace=True)
urgs_account.rename(columns={'urgs_application_status': 'letter_description'}, inplace=True)

# 3. Merge on pk_account to bring in district_description
merged = pd.merge(
    urgs_account, 
    account_district[['pk_account', 'district_description']],
    on='pk_account',
    how='left'
)

# 4. Rename amount and evt_datetime
merged.rename(columns={
    'amount': 'debt',
    'evt_datetime': 'urgs_date'
}, inplace=True)

# 5. Create the new DataFrame
new_df = merged[['district_description', 'pk_account', 'debt', 'urgs_date']]

# 6. Inspect the result
print("new_df shape:", new_df.shape)
display(new_df.head())
```

```python
# Drop NaN debts and convert dates
clean_df = new_df.dropna(subset=['debt']).copy()
clean_df['urgs_date'] = pd.to_datetime(clean_df['urgs_date'])

# Inspect
print("Cleaned DataFrame shape:", clean_df.shape)
print(clean_df.dtypes)
display(clean_df.head())
```

```python
print("=== clean_df info ===")
clean_df.info()

print("\n=== clean_df head ===")
display(clean_df.head())

print("\n=== clean_df describe ===")
display(clean_df.describe(include='all'))
```

```python
# Filter new_df for roles “O” or “T”
filtered_df = new_df[new_df['pk_porg_prop_role'].isin(['O', 'T'])]

# Optional: inspect the result
print(filtered_df.shape)
filtered_df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Copy filtered_df to plot_df
plot_df = filtered_df.copy()

# 2. Encode roles as 0/1
plot_df['role_code'] = plot_df['pk_porg_prop_role'].map({'O': 0, 'T': 1})
role_categories = ['O', 'T']

# 3. Create a discrete colormap with 2 colors
cmap = plt.cm.get_cmap('tab10', 2)

# 4. Scatter plot with color mapping
plt.figure()
scatter = plt.scatter(
    plot_df['urgs_date'],
    plot_df['debt'],
    c=plot_df['role_code'],
    cmap=cmap,
    s=50
)
plt.xlabel('URGS Date')
plt.ylabel('Debt')
plt.title('Debt vs URGS Date (O vs T)')

# 5. Colorbar with role labels
cbar = plt.colorbar(scatter, ticks=[0, 1])
cbar.ax.set_yticklabels(role_categories)
cbar.set_label('pk_porg_prop_role')

plt.tight_layout()
plt.show()
```
